```

def BFG1():
    w_min = 1000
    w_max = 10000
    t0 = 0
    t1 = 1
    h = 0.01
    freq = lambda t: w_min+alpha*t
    t_len = int((t1-t0)/h)
    alpha = (w_max-w_min)/t_len
    T=np.linspace(t0,t1,t_len)
    x = np.asarray([cos(t*freq(t)) for t in T])# добавить нули
    x_new = np.hstack((x, np.zeros(20*t_len)))
    y = np.fft.fft(x_new)
    
    XX=[]
    for i in range(0,len(y)):
        XX.append(i)
        
    return [[XX,np.abs(y)]]

def BFG2():
    result = BFG1()
    y = result[0][1]
    
    y_1 = y[:int(len(y)/2)]
    
    return [[ result[0][0], abs(y_1) ]]



def BFG3():
    
    w_min = 1000
    w_max = 10000
    t0 = 0
    t1 = 1
    h = 0.01
    freq = lambda t: w_min+alpha*t
    t_len = int((t1-t0)/h)
    alpha = (w_max-w_min)/t_len
    T=np.linspace(t0,t1,t_len)
    x = np.asarray([cos(t*freq(t)) for t in T])
    x_new = np.hstack((x, np.zeros(20*t_len)))
    
    XX=[]
    for i in range(0,len(x_new)):
        XX.append(i)
        
    X_prod = x_new * x_new
    res = np.fft.ifft(X_prod)
    
    return [[ XX, np.abs(res) ]]

def BFG4():
    result = BFG3()
    res = result[0][1]
    N_res = len(res)
    res_res = np.hstack((res[N_res//2:N_res], res[0:N_res//2]))
    
    
    XX=[]
    for i in range(0,len(res_res)):
        XX.append(i)
        
    return [[ XX,np.abs(res_res) ]]

def css_widget(text, font_size=10):
    html = "<h1 style='font-size:"+str(font_size)+"px; margin-left: 50px; margin-top: 0;margin-bottom:0;width:350px'>" + str(text) + "</h1>"
    W = widgets.HTML(value=html)
    return W

```
